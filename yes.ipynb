```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins
```

```python
df = pd.DataFrame(penguins)
sns.scatterplot(data=df, x= "bill_length_mm", y= "bill_depth_mm")
plt.title("relationship b/w two varibles")
plt.xlabel("jhab")
plt.ylabel("jhab")
```

```python
df = pd.DataFrame(penguins)

# Create a bar chart
sns.scatterplot(data=df, x= "bill_length_mm", y= "bill_depth_mm")

# Add labels and title
plt.title('Average between Species', fontsize=16)
plt.xlabel("bill_length_mm", fontsize=12)
plt.ylabel("bill_depth_mm", fontsize=12)

# Show the chart
plt.show()
```

```python
df = pd.DataFrame(penguins)

# Create a bar chart
sns.barplot(data=df, x='species', y='body_mass_g', palette='muted')
```

```python
df = pd.DataFrame(penguins)

# Create a bar chart
sns.barplot(data=df, x='species', y='body_mass_g', palette='muted')

plt.title('Average Body Mass by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Body Mass (g)', fontsize=12)

# Show the chart
plt.show()
```

```python
# Load the penguins dataset
data = sns.load_dataset("penguins")

# Create a box plot to compare male and female penguins
sns.boxplot(x="species", y="flipper_length_mm", hue="sex", data=data, palette="Set2")

# Add labels and title
plt.xlabel("Penguin Species")
plt.ylabel("Flipper Length (mm)")
plt.title("Box Plot of Flipper Length by Penguin Species and Sex")

# Show the plot
plt.show()
```

```python
# Load the penguins dataset
data = sns.load_dataset("penguins")

# Count the number of males and females
gender_counts = data["sex"].value_counts()

# Create a pie chart
plt.pie(
    gender_counts, 
    labels=gender_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=sns.color_palette("pastel")
)

# Add a title
plt.title("Proportion of Male and Female Penguins")

# Show the plot
plt.show()
```

```python
df = pd.DataFrame(penguins)

# Create a bar chart
sns.histplot(data=df, x= "island", y= "body_mass_g")

# Add labels and title
plt.title('Average between Species', fontsize=16)
plt.xlabel("island", fontsize=12)
plt.ylabel("body_mass_g", fontsize=12)

# Show the chart
plt.show()
```
